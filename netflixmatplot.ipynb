```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_titles.csv')
```

```python
df.head(3)
```

```python
df=df.dropna(subset=['type','release_year','rating','duration','country'])
type_counts=df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index, type_counts.values, color=["red","pink"])
plt.xlabel("movie or TV Show")
plt.ylabel("number of shows or movies")
plt.title("TV Show and Movies comparision")
plt.savefig('TV_vs_Movie.png')
plt.show()
```

```python
rating_counts=df['rating'].value_counts()
plt.pie(rating_counts.values, labels=rating_counts.index,autopct='%1.1f%%',startangle=90)
plt.title("Percentage of ratings")
plt.savefig('content_ratings.png')
plt.tight_layout()
plt.show()
```

```python
rating_counts
```

```python
movie_df=df[df['type']=='Movie'].copy()
```

```python
movie_df.head(5)
```

```python
movie_df['duration_int']=movie_df['duration'].str.replace('min','').astype(int)
plt.hist(movie_df['duration_int'],bins=30,color='purple',edgecolor='pink')
plt.title('duration of movie')
plt.xlabel('duration in minutes')
plt.ylabel('number of movies')
plt.tight_layout()
plt.show()
```

```python
release_year=movie_df['release_year'].value_counts()
plt.scatter(release_year.index,release_year.values, marker='o',color='red')
plt.title('release year vs number of Movies')
plt.xlabel('Release year')
plt.ylabel('number of shows')
plt.show()
```

```python
#top  ten countries with highest shows
countries=df['country'].value_counts().head(10)
countries
```

```python
plt.bar(countries.index,countries.values,color=['red','pink','blue','skyblue','purple','green','yellow','teal','orange','violet'])
plt.title("top ten countries")
plt.xlabel("countris")
plt.xticks(rotation=90)
plt.ylabel("number of shows")
plt.tight_layout()
plt.show()
```
